#Task

Write a Python program that calculates the area and perimeter of a rectangle. Define a function with a suitable name that takes the length and width of the rectangle as arguments and returns the area and perimeter. Use this function to calculate and print the properties of a rectangle with a length of 10 units and a width of 5 units.

Requirements:
Define a function calculate_rectangle_properties with two parameters: length and width.
Inside the function, calculate the area and perimeter of the rectangle using the formulas: area = length * width and perimeter = 2 * (length + width).
Return the calculated area and perimeter as a tuple.
Call the function with length = 10 and width = 5, and store the result in variables area and perimeter.
Print the calculated area and perimeter in a formatted string.
Ensure the code is clean and readable.

```python
#defining a function to calculate the area and perimeter of a rectangle

def calculate_rectangle_properties(length, width):
    area = length * width
    perimeter = 2 * (length + width)
    return area, perimeter
```

```python
#using the function to calculate the properties of a rectangle with width 5 and length 10

area, perimeter = calculate_rectangle_properties(10, 5)
print(f"The area of the rectangle is {area} units and the perimeter is {perimeter} units.")
#
```
