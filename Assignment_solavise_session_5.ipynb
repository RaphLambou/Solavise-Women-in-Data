##BMI Determination

```python
weight = float(input("Hello, I see you would like to know your BMI. \n Please enter your weight in kg:"))
```

```python
height = float(input("Now enter your height in m please"))
```

```python
bmi = weight / height ** 2
print(f'Great, your BMI is:{bmi}')
```

##Area of a triangle

```python
height = float(input("Enter the height of the triangle:"))
```
