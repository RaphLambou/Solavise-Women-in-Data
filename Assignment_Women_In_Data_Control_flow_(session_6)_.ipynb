#Task:
Write a Python program that:
1 . Asks the user to enter marks for Math, Science, and English.
Calculates the average score.
Prints:
"Excellent! You got an A" if average ≥ 90
"Very Good! You got a B" if average is between 80–89
"Good! You got a C" if average is between 70–79
"Needs Improvement! You got a D" if average is between 60–69
"Fail! You got an F" if average < 60
“Wow! You got a perfect score” if average = 100
“Oh no! You failed everything” if average = 0
“Invalid input. Try again!” if number is not in the range 0 - 100


##Basic code

```python
math = int(input("Enter your Math score: "))
science = int(input("Enter your Science score: "))
english = int(input("Enter your English score: "))

if (0 <= math <= 100 and 0 <= science <= 100 and 0 <= english <= 100):
  average = (math+science+english)/3
  if average >= 90:
    print("Excellent! You got an A")
  elif average >= 80:
    print("Very Good! You got a B")
  elif average >= 70:
    print("Good! You got a C")
  elif average >= 60:
    print("Needs Improvement! You got a D")
  else:
    print("Fail! You got an F")
else:
  print("Invalid input, try again")
```

##More sophisticated code

```python
scores = []
subjects = ['maths', 'sciences', 'english']

for subject in subjects:
  while True:
    score = int(input(f"Enter your {subject} score: "))
    if 0 <= score <= 100:
      scores.append(score)
      break  # Exit the inner while loop if input is valid
    else:
      print("Invalid input. Score must be between 0 and 100.")


average = sum(scores) / len(scores) # Calculate average after collecting all valid scores

if average == 100:
  print("Wow! You got a perfect score")
elif average == 0:
  print("Oh no! You failed everything")
elif average >= 90:
  print("Excellent! You got an A")
elif average >= 80:
  print("Very Good! You got a B")
elif average >= 70:
  print("Good! You got a C")
elif average >= 60:
  print("Needs Improvement! You got a D")
else:
  print("Fail! You got an F")
```
